```python
#genres
#id
#keywords
#overview
#popularity
#title
#vote_average
#vote_count
#cast
#crew

dataR = dataR[['genres','id','keywords','overview','popularity', 'title', 'vote_average', 'vote_count','cast', 'crew']]
```

```python
dataR.head()
```

```python
dataR.isnull().sum()
```

```python
dataR.dropna(inplace = True)
```

```python
dataR.isnull().sum()
```

```python
dataR.iloc[0].genres
```

```python
#Python literal_eval is a function defined in “ast” class of built-in class library.
#The “ast” expands to Abstract Syntax Tree. The ast module helps Python applications to process trees of the Python abstract syntax grammar.
def convert(obj):
  dataname = []
  for i in ast.literal_eval(obj):
    dataname.append(i['name'])
  return dataname
```

```python
dataR['genres']=dataR['genres'].apply(convert)
```

```python
dataR['keywords']=dataR['keywords'].apply(convert)
```

```python
dataR['cast'].iloc[0]
```

```python
def convert1(obj):
  dataname1 = []
  for i in ast.literal_eval(obj):
    if i['order']<3:
      dataname1.append(i['name'])
  return dataname1
```

```python
convert1(dataR['cast'].iloc[0])
```

```python
dataR['cast']=dataR['cast'].apply(convert1)
```

```python
def director(obj):
  list1 = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      list1.append(i['name'])
  return list1
```

```python
dataR['crew'] = dataR['crew'].apply(director)
```

```python
dataR.head()
```

```python
dataR['overview'] = dataR['overview'].apply(lambda x:x.split())
```

```python
dataR.head()
```

```python
#here there is a problem since we have space in between some names and tags it might be recommended us something different,
#eg if we want to watch sam worthinton movie it might us recommended us sam mendes movie.
#so to avoid such situation we should remove the space betwween our tags.

dataR['genres'] = dataR['genres'].apply(lambda x:[i.replace(" ","") for i in x ])
```

```python
dataR['keywords'] = dataR['keywords'].apply(lambda x:[i.replace(" ","") for i in x ])
dataR['cast'] = dataR['cast'].apply(lambda x:[i.replace(" ","") for i in x ])
dataR['crew'] = dataR['crew'].apply(lambda x:[i.replace(" ","") for i in x ])
```

```python
dataR.head()
```

```python
dataR['tags'] = dataR['overview'] + dataR['genres'] + dataR['keywords'] + dataR['cast'] + dataR['crew']
```

```python
dataR.head()
```

```python
req_data = dataR[['id', 'title' , 'tags']]
```

```python
req_data.head()
```

```python
req_data['tags'] = req_data['tags'].apply(lambda x:" ".join(x))
```

```python
req_data['tags'] = req_data['tags'].apply(lambda x:x.lower())
```

```python
req_data.head()
```

```python
# now we will convert each tags into some vectors and each vector will represent one movie also the vector
#which is close to one another will be recommended means closeness of the vector will give us the similarity between the movies
#we use 5000 words taken form each movie to make 1 vector for each movie

```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
  l = []
  for i in text.split():
    r = ps.stem(i)
    l.append(r)
  return " ".join(l)
```

```python
req_data['tags'] = req_data['tags'].apply(stem)
```

```python
#from sklearn.feature_extraction.text import CountVectorizer
#cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(max_features = 5000, stop_words = 'english')
```

```python
Vector = tfv.fit_transform(req_data['tags']).toarray()
```

```python
Vector.shape #4806 movies and 5000 words
#since there are not that much atgs for each movies so there are much empty blocks in the matrix
```

```python
tfv.get_feature_names()
```

```python
#see now we have tags that have words like cook cooking thats means the same so we will apply stanning that will make the word to cook cook
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarty = cosine_similarity(Vector)
```

```python
similarty.shape #similarty measure with each movive with each other movie
```

```python
sorted(enumerate(similarty[0]),reverse=True, key = lambda x:x[1])[1:6]
```

```python
def recommended(movie):
  index = req_data[req_data['title'] == movie].index[0]
  similar = similarty[index]
  req_movies = sorted(enumerate(similar),reverse=True, key = lambda x:x[1])[1:6]
  for i in req_movies:
    print(req_data['title'].iloc[i[0]])
```

```python
recommended('Avatar')
```
